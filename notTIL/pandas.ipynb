### Filtering & Sorting

```python
# 20. 데이터를 로드하라.
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df2 = pd.read_csv(DataUrl)
```

```python
# 21. quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라
df2[df2['quantity']==3].head(5)
```

```python
# 22. quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라
df2[df2['quantity']==3].reset_index().head(5)
```

```python
# 23. quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라
df2[['quantity','item_price']]
```

```python
# 24.item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라
import re
df2['new_price'] = df2['item_price'].apply(lambda x : re.sub('\$','',x)).astype('float64')
df2
```

```python
# 25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라
df2_1 = df2[df2['new_price']<= 5]
len(df2_1)
```

```python
# 26. item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하고 index 값을 초기화 하여라
df2_2 = df2[df2['item_name']== 'Chicken Salad Bowl'].reset_index()
df2_2
```

```python
# 27. new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라
df2[(df2['new_price']<=9) & (df2['item_name']=='Chicken Salad Bowl')].head(5)
```

```python
# 28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라
df2_3 = df2.sort_values(by='new_price').reset_index()
```

```python
# 29. df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라
df2[df2['item_name'].str.contains('Chips')]
```

```python
# 30. df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라
df2.iloc[:,::2]
```

```python
# 31. df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라
df2_4 = df2.sort_values(by='new_price', ascending=False).reset_index()
df2_4
```

```python
# 32. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱하라
df2[(df2['item_name']=='Steak Salad')|(df2['item_name']=='Bowl')]
```

```python
# 33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라
df2_5 = df2[(df2['item_name']=='Steak Salad')|(df2['item_name']=='Bowl')]
df2_5[~df2_5['item_name'].duplicated(keep='first')]
# df2_5.drop_duplicates(subset='item_name',keep='first')
```

```python
# 34. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라
df2_5[~df2_5['item_name'].duplicated(keep='last')]
# df2_5.drop_duplicates(subset='item_name',keep='last')
```

```python
# 35. df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱하라

df2.loc[df2['new_price']>=df2['new_price'].mean()]
```

```python
# 37. df의 데이터 중 choice_description 값이 NaN 인 데이터의 갯수를 구하여라
df2['choice_description'].isnull().sum()
```

```python
# 38. df의 데이터 중 choice_description 값이 NaN 인 데이터를 NoData 값으로 대체하라(loc 이용)
df2.loc[df2['choice_description'].isnull(),'choice_description'] = 'NoData'
df2
```

```python
# 39. df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 인덱싱하라
df2.loc[df2['choice_description'].str.contains('Black')]
```

```python
# 40. df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수를 출력하라
len(df2.loc[~df2['choice_description'].str.contains('Vegetables')])
```

```python
# 41. df의 데이터 중 item_name 값이 N으로 시작하는 데이터를 모두 추출하라
df2.loc[df2['item_name'].str.startswith('N')].head(3)
```

```python
# 42. df의 데이터 중 item_name 값의 글자갯수가 15개 이상인 데이터를 인덱싱하라
df2.loc[df2['item_name'].str.len() >= 15]
```

```python
# 43. df의 데이터 중 new_price값이 lst에 해당하는 경우의 데이터 프레임을 구하고 그 갯수를 출력하라 
lst =[1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]
df2.loc[df2['new_price'].isin(lst)]
```
